```python
import pandas as pd
from utils.config import load_config
from utils.scoring import get_trained_model_by_name, score_model
```

```python
config = load_config()
```

```python
for model_name in config.models:
    for data_dim in config.sample_sizes:

        print(
            f"Start scoring {model_name} model for {data_dim}x{config.test_samples_num} data..."
        )

        model = get_trained_model_by_name(model_name, data_dim)
        data = pd.read_csv(
            f"{config.data_dir}/{config.test_data_subdir}/fbm_{data_dim}x{config.test_samples_num}.csv"
        )

        _ = score_model(
            experiment_name=config.scoring_experiment_name,
            run_name=f"{model_name}_{data_dim}",
            model=model,
            X_test=data.drop(columns=[config.target_column]),
            y_test=data[config.target_column],
        )
```
